```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
        
    return data
```

```python
data = load_data("abv.json")
```

```python
print(data)
print(type(data))
```

```python
def clean_data(data):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    
    unique_num=set()
    cleaned_data=[]
    for user in data:
        raw_rat=str(user["rating"]).strip().lower()
        if raw_rat in text_to_num:
            user["rating"] = text_to_num[raw_rat]
        
        # 2️⃣ Convert digits directly ("1", "2", "5")
        elif raw_rat.isdigit():
            user["rating"] = int(raw_rat)

        raw_age=user.get("age")
        if(raw_age == None):
            user["age"]= None

        if (user["name"].strip() in unique_num):
            continue
              
        unique_num.add(user["name"])
        cleaned_data.append(user)


    return cleaned_data
```

```python
data=clean_data(data)
```

```python
def get_rating(new_data):
    tot_rat=0

    for user in new_data:
        tot_rat+=float(user["rating"])

    avg_rat=(tot_rat/len(data))
    print(avg_rat)
    poor_rating =0 
    for user in new_data:
        if (float(user["rating"])<3):
            poor_rating +=1
    print(poor_rating/len(data)*100)

```

```python
get_rating(data)
```
