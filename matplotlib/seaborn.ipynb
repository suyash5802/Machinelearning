```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
```

```python
tips=sns.load_dataset("tips")
print(type(tips))
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    kind="line"
)
```

```python
days=["Mon","Tues","Wedn","Thurs"]

city=["NY","TKY","NDLS","LDN"]

temp=[
    [10,22,34,56],
    [11,23,35,57],
    [13,24,36,58],
    [14,25,38,59]
]

fig, ax=plt.subplots(2,2)
cityno=0

for i in range(2):
    for j in range(2):
        ax[i][j].plot(days,temp[cityno])
        ax[i][j].set_title(city[cityno])
        ax[i][j].grid(True)
        cityno=cityno+1

fig.suptitle("temp")

fig.tight_layout()
```

```python
fig, ax = plt.subplots() 
sns.boxplot(data=sns.load_dataset("tips"), x="day", y="total_bill", hue="sex", ax=ax)
ax.set_title("Boxplot: Total Bill by Day and Sex")
ax.set_xlabel("Day of the Week") 
ax.set_ylabel("Total Bill ($)") 
plt.legend(title="Sex") 
plt.show()
```
