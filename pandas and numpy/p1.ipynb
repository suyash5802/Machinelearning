```python
df=pd.read_csv("data.csv")

df.head(5)
df.shape
df.rename(columns={"name":"Name"})
```

```python
df[df["city"]=="Delhi"]
```

```python
pdf= df[(df["category"]=="Electronics") &  (df["city"]=="Delhi")] 
pdf[["city","customer","product","price"]]
```

```python
df["total_amount"]=df[["price","quantity"]].apply(lambda x:x.quantity*x.price , axis=1)
df["total_amount"].sum()
df.groupby("city")["total_amount"].mean()
df.groupby("product")["quantity"].count()
df.groupby("customer")["total_amount"].sum().sort_values(ascending=False)

df.groupby("product")["total_amount"].sum().nlargest(2)
df.groupby("city")["total_amount"].sum()

```
