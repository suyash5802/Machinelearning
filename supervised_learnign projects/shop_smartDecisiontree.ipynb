```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("shop_smart_ecommerce.csv")
df.info()
df
```

```python
features=["ProductRelated","ProductRelated_Duration","PageValues","SpecialDay","Region","VisitorType","Weekend"]
target=["Revenue"]
```

```python
df
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

ohe = OneHotEncoder(sparse_output=False, drop='first')
week_enc=ohe.fit_transform(df[["Weekend"]])
week_df=pd.DataFrame(
    week_enc,columns=ohe.get_feature_names_out(['Weekend']),
    index=df.index

)
df = pd.concat([df.drop('Weekend', axis=1), week_df], axis=1)

```

```python
X=df[features]
y=df[target]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state=42
)
```
