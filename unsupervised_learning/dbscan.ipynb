```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
lbc=load_iris()
X=lbc.data
y=lbc.target
```

```python
ss=StandardScaler()
X_scaled=ss.fit_transform(X)
sns.scatterplot(x=X[:,0],y=X[:,1],c=y)
```

```python
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=0.4,min_samples=6)
label=model.fit_predict(X_scaled)
sns.scatterplot(x=X[:,0],y=X[:,2],c=label)
```

```python
X,y=make_moons(n_samples=1000,noise=0.1,random_state=42)

ss=StandardScaler()
X_scaled=ss.fit_transform(X)
sns.scatterplot(x=X[:,0],y=X[:,1])
```

```python
model=DBSCAN(min_samples=5,eps=0.2)
label=model.fit_predict(X_scaled)
sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,1],c=label)
```
